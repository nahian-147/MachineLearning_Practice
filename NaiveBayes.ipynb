```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = ['sweet','long','yellow','banana']
f2 = ['sweet','round','yellow','orange']
```

```python
total_fruits = 1000
```

```python
class fruit:
    sweet = 0
    sour = 0
    long = 0
    not_long = 0
    yellow = 0
    green = 0
    total = 0
    name = ''
```

```python
b = fruit()
b.sweet = 380
b.sour = 20
b.long = 400
b.not_long = 0
b.yellow = 350
b.green = 50
b.total = 400
b.name = 'banana'
```

```python
o = fruit()
o.sweet = 210
o.sour = 90
o.long = 0
o.not_long = 300
o.yellow = 200
o.green = 100
o.total = 300
o.name = 'orange'
```

```python
a = fruit()
a.sweet = 170
a.sour = 30
a.long = 0
a.not_long = 200
a.yellow = 0
a.green = 200
a.total = 200
a.name = 'apples'
```

```python
g = fruit()
g.sweet = 85
g.sour = 15
g.long = 40
g.not_long = 60
g.yellow = 10
g.green = 90
g.total = 100
g.name = 'grapes'
```

```python
total_sweet = b.sweet+o.sweet+a.sweet+g.sweet
total_sour = b.sour+o.sour+a.sour+g.sour
total_long = b.long+o.long+a.long+g.long
total_round = b.not_long+o.not_long+a.not_long+g.not_long
total_yellow = b.yellow+o.yellow+a.yellow+g.yellow
total_green = b.green+o.green+a.green+g.green
```

```python
fruits = {'banana':[],'orange':[],'apples':[],'grapes':[]}
```

```python
def probabilities(fruit):
    fruits[fruit.name].append(fruit.sweet/fruit.total)
    fruits[fruit.name].append(fruit.sour/fruit.total)
    fruits[fruit.name].append(fruit.long/fruit.total)
    fruits[fruit.name].append(fruit.not_long/fruit.total)
    fruits[fruit.name].append(fruit.yellow/fruit.total)
    fruits[fruit.name].append(fruit.green/fruit.total)
    fruits[fruit.name].append(fruit.total)
```

```python
probabilities(b)
probabilities(o)
probabilities(a)
probabilities(g)
fruits
prob = np.array([total_sweet,total_sour,total_long,total_round,total_yellow,total_green])/total_fruits
```

```python
def fruit_probability(c):
    p = []
    for f in fruits:
        P = 1
        for _ in c:
            P *= fruits[f][_]/prob[_]
        P += fruits[f][-1]/total_fruits
        p.append(P)
    return p
```

```python
fruit_probability([1,3,4])
```

```python
plt.figure(figsize=(10,5),dpi=200)
plt.bar(['banana','orange','apple','grape'],fruit_probability([0,3,5]),width=.2,color='green')
```
